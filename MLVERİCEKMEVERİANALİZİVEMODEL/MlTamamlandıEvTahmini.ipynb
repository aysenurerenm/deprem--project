```python
#ilan güncelleme tarihi sütununu tarih formatına çevirdim
import pandas as pd
import unicodedata

# 1) Tüm görünmez karakterleri temizle
def clean_text(x):
    if pd.isna(x):
        return x
    x = str(x)
    x = unicodedata.normalize("NFKD", x)     # Unicode normalize
    x = x.replace("\u200f", "")              # sağdan sola işaretlerini sil
    x = x.replace("\xa0", " ")               # nbsp temizle
    return x.strip()

df_merged2["İlan Güncelleme Tarihi"] = df_merged2["İlan Güncelleme Tarihi"].apply(clean_text)

# 2) Türkçe → İngilizce ay çevir
months = {
    "Ocak": "January",
    "Şubat": "February",
    "Mart": "March",
    "Nisan": "April",
    "Mayıs": "May",
    "Haziran": "June",
    "Temmuz": "July",
    "Ağustos": "August",
    "Eylül": "September",
    "Ekim": "October",
    "Kasım": "November",
    "Aralık": "December"
}

for tr, en in months.items():
    df_merged2["İlan Güncelleme Tarihi"] = df_merged2["İlan Güncelleme Tarihi"].str.replace(tr, en, regex=False)


df_merged2["İlan Güncelleme Tarihi"] = pd.to_datetime(
    df_merged2["İlan Güncelleme Tarihi"],
    dayfirst=True,
    errors="coerce"
)


df_merged2["İlan Güncelleme Tarihi"].head()
```

```python
age_mapping = {#sözlük oluşturduk
    '0 (Yeni)': 0,
    '1': 1,
    '2': 2,
    '3': 3,
    '4': 4,
    '5-10': 7,       # aralık ortalaması
    '11-15': 13,
    '16-20': 18,
    '21 Ve Üzeri': 25  # üst sınır, ihtiyaca göre değişebilir
}

# 1) Dönüştürme
df_merged2['Binanın Yaşı Num'] = df_merged2['Binanın Yaşı'].replace(age_mapping)#dönüştürme


df_merged2['Binanın Yaşı Num'] = df_merged2['Binanın Yaşı Num'].astype(float)#Eksik değer varsa float olarak bırak
df_merged2['Binanın Yaşı Num'].head()#kontrol
```

```python
import pandas as pd

# 0 olması gereken kat tipleri, 0 olacaklar
ozel_katlar_sifir = [
    "Yüksek Giriş", "Düz Giriş (Zemin)", "Bahçe Katı", 
    "Bahçe Dublex", "Villa Tipi"
]

# Kot tipleri, negatif olacaklar
kot_katlar = [
    "Kot 2 (-2).Kat", "Kot 3 (-3).Kat", "Bodrum Kat"
]

# 1) Sayısal kısmı al (pozitif ve negatif sayıları yakalar)
df_merged2['Bulunduğu Kat Num'] = df_merged2['Bulunduğu Kat'].str.extract(r'(-?\d+)')

# 2) Sayıya çevir, eksik değerleri 0 yap
df_merged2['Bulunduğu Kat Num'] = df_merged2['Bulunduğu Kat Num'].fillna(0).astype(int)

# 3) Özel katlar (0 olacaklar)
df_merged2.loc[df_merged2['Bulunduğu Kat'].isin(ozel_katlar_sifir), 'Bulunduğu Kat Num'] = 0

# 4) Kot katlar (negatif olacaklar)
for kot in kot_katlar:
    match = df_merged2['Bulunduğu Kat'] == kot
    if '(' in kot:
        # Örn: "Kot 2 (-2).Kat" -> -2
        neg_value = int(kot.split('(-')[1].split(')')[0]) * -1
    else:
        # Bodrum Kat -> -1 olarak atayabiliriz
        neg_value = -1
    df_merged2.loc[match, 'Bulunduğu Kat Num'] = neg_value

# 5) Çatı Katı → bina kat sayısına eşitle
if 'Binanın Toplam Kat Sayısı' in df_merged2.columns:
    df_merged2.loc[df_merged2['Bulunduğu Kat'] == "Çatı Katı", 'Bulunduğu Kat Num'] = df_merged2['Binanın Toplam Kat Sayısı']
else:
    # Eğer yoksa, maksimum kat sayısı ile eşitle
    max_kat = df_merged2['Bulunduğu Kat Num'].max()
    df_merged2.loc[df_merged2['Bulunduğu Kat'] == "Çatı Katı", 'Bulunduğu Kat Num'] = max_kat

# 6) Kontrol
print(df_merged2['Bulunduğu Kat Num'].value_counts().sort_index())

```

```python
import numpy as np
import pandas as pd

def parse_oda(value):
    value = str(value).strip()
    
    # Stüdyo → salon=1, oda=0
    if value.lower() == "stüdyo":
        return 1, 0
    
    # "1 Oda" → salon=0, oda=1
    if "oda" in value.lower() and "+" not in value:
        num = float(value.split()[0].replace(",", "."))
        return 0, num
    
    # "X+Y" formatı → X oda, Y salon
    if "+" in value:
        parts = value.split("+")
        try:
            oda = float(parts[0].replace(",", ".").strip())
            salon = float(parts[1].replace(",", ".").strip().split()[0])
            return salon, oda
        except:
            return np.nan, np.nan
    
    # Tek sayı → salon=0, oda=value
    try:
        num = float(value.replace(",", "."))
        return 0, num
    except:
        return np.nan, np.nan

# Uygulama
df_merged2[['Salon Sayısı', 'Oda Sayısı Num']] = df_merged2['Oda Sayısı'].apply(lambda x: pd.Series(parse_oda(x)))

# Toplam oda sayısı
df_merged2['Toplam Oda Sayısı'] = df_merged2['Salon Sayısı'] + df_merged2['Oda Sayısı Num']

df_merged2[['Oda Sayısı', 'Salon Sayısı', 'Oda Sayısı Num', 'Toplam Oda Sayısı']].head(15)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Özellikler ve hedef
X = df_modeleHazir.drop(columns=["Fiyat"])
y = df_modeleHazir["Fiyat"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=44)

# Random Forest parametreleri
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Gradient Boosting parametreleri
gb_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 1.0]
}

# RandomizedSearchCV
rf_random = RandomizedSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_distributions=rf_param_grid,
    n_iter=20,  # 20 farklı kombinasyon dene
    cv=3,
    verbose=2,
    n_jobs=-1,
    scoring='r2'
)

gb_random = RandomizedSearchCV(
    estimator=GradientBoostingRegressor(random_state=42),
    param_distributions=gb_param_grid,
    n_iter=20,
    cv=3,
    verbose=2,
    n_jobs=-1,
    scoring='r2'
)

# Fit
rf_random.fit(X_train, y_train)
gb_random.fit(X_train, y_train)

# En iyi parametreler
print("Random Forest Best Params:", rf_random.best_params_)
print("Gradient Boosting Best Params:", gb_random.best_params_)

# En iyi modellerle tahmin
rf_best = rf_random.best_estimator_
gb_best = gb_random.best_estimator_

y_pred_rf = rf_best.predict(X_test)
y_pred_gb = gb_best.predict(X_test)

print("Random Forest R2:", r2_score(y_test, y_pred_rf))
print("Gradient Boosting R2:", r2_score(y_test, y_pred_gb))
print("Gradient Boosting R2:", r2_score(y_test, y_pred_gb))
print("Gradient Boosting RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_gb)))
```
